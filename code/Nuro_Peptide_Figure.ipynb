```python
df_tr = pd.read_csv(r'C:\Users\tareq\Desktop\Thesis\DataSet\Combine_Data\AAC.csv')
columns = df_tr.columns.tolist()
# Filter the columns to remove data we do not want
columns = [c for c in columns if c not in ["Target"]]
# Store the variable we are predicting
target = "Target"
X = df_tr[columns]
Y = df_tr[target]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size= 0.3)
```

```python
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
for train_i, test_i in cv.split(xtrain):
    X_train1, X_test1 = xtrain.iloc[train_i], xtrain.iloc[test_i]
    y_train1, y_test1 = ytrain.iloc[train_i], ytrain.iloc[test_i]
```

```python
rf = RandomForestClassifier()
dt = DecisionTreeClassifier()
etc = ExtraTreesClassifier()
knn = KNeighborsClassifier()
cat = CatBoostClassifier()
adb = AdaBoostClassifier()
xgb = XGBClassifier()
lgb = LGBMClassifier()
```
